# Interpolation

```python
import matplotlib.pyplot as plt
import numpy as np 
plt.rcParams['font.size'] = 15
```

```python
def func(x):
    y = np.sin(np.pi*x/6)
    return y 

```

```python

x_pts = np.array([-1, 1, 3, 5, 8 ])
y_pts = func(x_pts)

x_vals = np.arange(np.min(x_pts)-2, np.max(x_pts)+2 ,0.1)
y_vals = func(x_vals)

plt.plot(x_vals,y_vals,ls='--',lw=3,color='r',label=r'$f(x)$')
plt.plot(x_pts,y_pts,ls='',marker= 'o',ms=10,lw=2,color='r',markeredgecolor='k')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

```

```python
def lagrange(x,x_pts,y_pts):
    N = len(x_pts)
    #######################
    def L(x,i):
        prod = 1
        for j in range(N):
            if j != i:
                prod = prod*(x-x_pts[j])/(x_pts[i]-x_pts[j])
        return prod
    ########################
    
    P = 0 # value of P(x) at x 
    for i in range(N):
        P = P + y_pts[i]*L(x,i)
    return P 


P_vals = np.zeros_like(x_vals)
for i in range(len(P_vals)):
    P_vals[i] = lagrange(x_vals[i],x_pts,y_pts)
```

```python
plt.plot(x_vals,y_vals,ls='--',lw=3,color='red',label=r'$f(x)$')
plt.plot(x_vals,P_vals,ls='-',lw=2,color='blue', label=r'$P(x)$')
plt.plot(x_pts,y_pts,ls='',marker= 'o',ms=10,lw=2,color='r',markeredgecolor='k')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
